## Evaluation with human-in-the-loop

For some hard problems that these LLM applications cannot solve automatically, we want to test whether human feedback can be helpful during the solving process. 

Setup:
We test one problem on MathChat, Auto-GPT and ChatGPT+Plugin, which could incorporate human feedback during the solving process.
We do not change any initial prompt or setting, (for example, including phrase like "with human feedback"). 

-------------
**Problem**: Let $a_1,$ $a_2,$ $\dots$ be a sequence of real numbers such that for all positive integers $n,$
$$\sum_{k = 1}^n a_k \left( \frac{k}{n} \right)^2 = 1.$$
Find the smallest $n$ such that $a_n < \frac{1}{2018}.$

**Solution**: For $n = 1,$ we get $a_1 = 1.$  Otherwise,
\[\sum_{k = 1}^n k^2 a_k = n^2.\]Also,
\[\sum_{k = 1}^{n - 1} k^2 a_k = (n - 1)^2.\]Subtracting these equations, we get
\[n^2 a_n = n^2 - (n - 1)^2 = 2n - 1,\]so $a_n = \frac{2n - 1}{n^2} = \frac{2}{n} - \frac{1}{n^2}.$  Note that $a_n = 1 - \frac{n^2 - 2n + 1}{n^2} = 1 - \left( \frac{n - 1}{n} \right)^2$ is a decreasing function of $n.$ Also,
\[a_{4035} - \frac{1}{2018} = \frac{2}{4035} - \frac{1}{4035^2} - \frac{1}{2018} = \frac{1}{4035 \cdot 2018} - \frac{1}{4035^2} > 0,\]and
\[a_{4036} < \frac{2}{4036} = \frac{1}{2018}.\]Thus, the smallest such $n$ is $\boxed{4036}.$


**Key Step**: The key step of this problem is: 1. move n^2 to the right of the equation. 2. Use the difference between $a_n$ and $a_(n-1)$ to get expression of $a_n$ with the only variable n.

----------------------

We follow this process:
1. Input the problem.
2. Observe the response (Always wrong). We give the first hint, which is to move n^2 to the right of the equation.
3. Communicate with the model to get the desired result:$ \sum_{k=1}^{n} k^2 a_k = n^2 $
4. Give the next hint: "use a_n - a_n-1 to solve this"
5. Communicate with the model to get the desired result: $a_n = \frac{n^2 - (n-1)^2}{n^2} = \frac{2n-1}{n^2}$. 
6. Ask the model to solve the rest. 


With our evaluation, MathChat and ChatGPT+Plugin can solve the problem correctly with human feedback, they are similar in terms of verbosity and follow human feedback. However, Auto-GPT rely on python execution to folow the first hint and we cannot get the desired output. We give up after a few rounds.
