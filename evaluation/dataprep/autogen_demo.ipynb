```python
from flaml import oai

config_list = oai.config_list_openai_aoai(openai_api_key_file='key_openai.txt')
```

```python
from flaml.autogen.agent import AssistantAgent, UserProxyAgent

# create an AssistantAgent instance named "assistant"
assistant = AssistantAgent(
    name="assistant",
    request_timeout=600,
    seed=42,
    config_list=config_list,
    model="gpt-4",  # modify if the endpoint you use doesn't support this model
    temperature=0,
)
# create a UserProxyAgent instance named "user"
user = UserProxyAgent(
    name="user",
    human_input_mode="TERMINATE",
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
    work_dir="./",
    use_docker=False,
)
```

```python
assistant.reset()
```

```python
assistant.receive("Open the file './Sport car price.csv', find out the average price of Ferrari.", user)
```

```python
assistant.receive("What is the average price of Mercades AMG?", user)
```
